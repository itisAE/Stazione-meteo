```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
# Importare le librerie necessarie
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_auc_score, roc_curve
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
dataset = pd.read_csv("datasetPulito_dd.csv")
```

## Pressione prevista domani

```python
dataset['Pressione prevista dopodopodomani'] = dataset['Pressione prevista dopodomani'].shift(-1)
dataset['Precipitazione prevista dopodopodomani'] = dataset['Precipitazione prevista dopodomani'].shift(-1)
dataset['Temperatura prevista dopodopodomani'] = dataset['Temperatura prevista dopodomani'].shift(-1)
dataset['Umidità prevista dopodopodomani'] = dataset['Umidità prevista dopodomani'].shift(-1)
dataset['Vento previsto dopodopodomani'] = dataset['Vento previsto dopodomani'].shift(-1)
```

```python
dataset = dataset.dropna()
```

```python
X = dataset.drop(columns = ['DATA', 'Pressione prevista dopodopodomani'])
y = dataset['Pressione prevista dopodopodomani'] #target

# Dividere il dataset in training e test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creare il modello di regressione lineare
modelPress = LinearRegression()
modelPress.fit(X_train, y_train)

# Fare predizioni sul test set
y_pred = modelPress.predict(X_test)


print("R^2 Score:", r2_score(y_test, y_pred))

# Visualizzare i risultati (grafico)
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='#f49055', alpha=0.5, label="Dati Reali vs Predetti")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='#692900', linestyle='--', label="Perfetta corrispondenza")
plt.xlabel("Pressione effettiva")
plt.ylabel("Pressione predetta")
plt.title("Regressione Pressione domani")
plt.legend()
plt.savefig('regressionePressioneDopodomani.png', dpi = 180, bbox_inches = 'tight')
plt.show()
```

```python
with open('pressione_prev_ddd.pkl', 'wb') as f:
    pickle.dump(modelPress, f) 
```
